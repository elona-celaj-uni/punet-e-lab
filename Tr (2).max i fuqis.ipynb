### 1. Qëllimi i Eksperimentit
Ky eksperiment synon të verifikojë teorikisht dhe eksperimentalisht kushtin e transferimit maksimal të fuqisë nga një burim me rezistencë të brendshme (Rth) te një ngarkesë e ndryshueshme (RL). Do të vërtetojmë që fuqia maksimale arrihet kur $R_L = R_{th}$. Përmes matjeve të tensionit në ngarkesë, do të llogarisim fuqinë për vlera të ndryshme rezistencash dhe do të ndërtojmë kurbën karakteristike.

### 2. Baza Teorike
Sipas Teoremës së Teveninit, çdo qark linear mund të zëvendësohet me një burim tensioni $V_{th}$ në seri me një rezistencë $R_{th}$.
Fuqia e transferuar te ngarkesa $R_L$ llogaritet si:
$$P_L = I^2 \cdot R_L = \left( \frac{V_{th}}{R_{th} + R_L} \right)^2 \cdot R_L$$

**Vlerat e projektit:**
* $V_{th} = 5.68 \text{ V}$
* $R_{th} = 11.03 \text{ k}\Omega$ (ose $\Omega$, sipas specifikimit tuaj)

```python
import pandas as pd
import numpy as np

# Parametrat e Teveninit
Vth = 5.68
Rth = 11.03  # kOhm

# Rezistencat e ngarkesës (RL) nga eksperimenti juaj
RL_values = [1.2, 2.7, 5.0, 6.8, 21.0, 58.0, 74.0, 81.0, 100.0] # në kOhm

# Krijimi i tabelës
df = pd.DataFrame({"RL (kOhm)": RL_values})

# Llogaritja e Tensionit në ngarkesë (V_L = Vth * RL / (Rth + RL))
df["VL (V)"] = Vth * df["RL (kOhm)"] / (Rth + df["RL (kOhm)"])

# Llogaritja e Fuqisë (P = VL^2 / RL) në mW
df["PL (mW)"] = (df["VL (V)"]**2 / df["RL (kOhm)"])

df.sort_values("RL (kOhm)", inplace=True)
df
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8,5))
plt.plot(df["RL (kOhm)"], df["PL (mW)"], 'ro-', label='Fuqia e matur')
plt.axvline(x=Rth, color='g', linestyle='--', label=f'Rth = {Rth} kOhm')

plt.xlabel("Rezistenca e Ngarkesës RL (kOhm)")
plt.ylabel("Fuqia PL (mW)")
plt.title("Grafiku i Transferimit të Fuqisë")
plt.legend()
plt.grid(True)
plt.show()
```

### 6. Analiza e Gabimeve dhe Diskutimi
Nga grafiku dhe tabela vërehet se fuqia rritet me rritjen e $R_L$ derisa i afrohet vlerës së $R_{th}$, dhe pastaj fillon të bjerë. 

**Burimet e gabimit:**
* **Toleranca e rezistorëve:** Vlerat nominale (p.sh. 5k$\Omega$) mund të kenë devijim $\pm 5\%$.
* **Efektet parazitare:** Rezistenca e kabllove lidhës mund të rrisë vlerën efektive të $R_{th}$.
* **Pasiguria e instrumentit:** Multimetri ka një gabim sistematik në matjen e tensioneve të vogla.

### 8. Përfundimi
Eksperimenti vërtetoi se transferimi maksimal i fuqisë ndodh kur rezistenca e ngarkesës është e barabartë me rezistencën e Teveninit të qarkut. 
* Vlera teorike për fuqinë maksimale: $P_{max} = \frac{V_{th}^2}{4 R_{th}}$.
* Devijimi midis pikës kulmore eksperimentale dhe vlerës $R_L = R_{th}$ ishte minimal, duke konfirmuar vlefshmërinë e modelit të Teveninit.

## EKSPERIMENTI II: Transferimi Maksimal i Fuqisë

### 1. Qëllimi dhe Modeli i Qarkut
Në këtë pjesë do të vërtetojmë se fuqia maksimale në ngarkesë arrihet kur $R_L = R_{th}$.
Qarku kompleks është thjeshtuar në një model ekuivalent të Teveninit me parametrat:
* **Vth (Vlera Open Source):** 5.68 V
* **Rth (Rezistenca Tevenin):** 11.03 kΩ

### 2. Të Dhënat Eksperimentale
Për vlera të ndryshme të $R_L$, kemi matur rrymën $I$ në qark

```python
import pandas as pd
import matplotlib.pyplot as plt


data_2 = {
    'RL (kOhm)': [2.25, 3.24, 6.82, 11.45, 15.38, 21.78, 37.7, 210.0, 301.0],
    'I_matur (mA)': [1.95, 1.9, 1.8, 1.6, 1.3, 1.2, 1.1, 1.05, 1.0]
}

df2 = pd.DataFrame(data_2)

# Llogaritja e fuqisë: P = I^2 * RL (në mWatt)
df2['P_matur (mW)'] = (df2['I_matur (mA)']**2) * df2['RL (kOhm)']

# Shfaqja e tabelës në Jupyter
print("Tabela e llogaritjeve për Eksperimentin 2:")
df2
```

```python
plt.figure(figsize=(9, 5))
plt.plot(df2['RL (kOhm)'], df2['P_matur (mW)'], 'ro-', label='Fuqia e llogaritur nga matjet')

# Vija vertikale që tregon ku duhet të ishte piku teorik (Rth)
plt.axvline(x=11.03, color='green', linestyle='--', label='Rth Teorike (11.03 kOhm)')

plt.title('Grafiku i Transferimit të Fuqisë (P vs RL)')
plt.xlabel('Rezistenca e Ngarkesës RL (kOhm)')
plt.ylabel('Fuqia PL (mW)')
plt.legend()
plt.grid(True, which="both", ls="-")
plt.show()
```

Burimet kryesore të gabimit
* **Tolerancat e rezistorëve:** Rezistorët realë kanë tolerancë (±1% ose ±5%), prandaj vlerat reale ndryshojnë nga ato nominale.
* **Pasiguritë e matjes:** Multimetri ka kufizim në saktësi dhe rezolucion.
* **Kufizimet e instrumenteve:** Rezistenca e brendshme e instrumenteve mund të ndikojë në matje.
* **Efektet parazitare:** Lidhjet dhe kontaktet krijojnë rezistenca shtesë të vogla.

### 4. Përfundimi i Përgjithshëm
Eksperimenti vërtetoi dy parime themelore të elektronikës:
1. Shpërndarjen e tensionit në një qark seri (Ndarësi).
2. Kushtin e transferimit maksimal të fuqisë kur $R_L \approx R_{th}$. Grafiku tregon qartë rritjen e fuqisë drejt vlerës 11.03 kΩ dhe rënien e saj pas kësaj pike.
   
   **Analiza e Grafikut:** 
Lakora e ndërtuar tregon sjelljen dinamike të qarkut: kur rezistenca e ngarkesës $R_L$ është e vogël, 
rryma është e lartë por tensioni është i ulët, duke dhënë fuqi të vogël. Kur $R_L$ rritet tej mase, 
tensioni është i lartë por rryma bie shumë, duke ulur përsëri fuqinë. 
Maksimumi i dukshëm në grafik vërteton se transferimi optimal ndodh vetëm në pikën e 
përshtatjes së rezistencave ($R_L = R_{th} = 11.03 \text{ k}\Omega$).
